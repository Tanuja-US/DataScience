# **Image** **to** **pencil** **sketch**

### **Importing required libraries**

```python
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow
```

### **Reading the image in RBG format**

```python
img = cv2.imread("Eagle.jpg") 
```

### **Displaying the image**

```python
cv2_imshow(img) 
```

### **Converting it to a grayscale image**

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img) 
```

### **Inverting the image (Negative image)**

```python
inverted_img = 255 - gray_img
cv2_imshow(inverted_img) 
```

### **Blur the image**

```python
blur_img = cv2.GaussianBlur(inverted_img, (21, 21), 0)
cv2_imshow(blur_img) 
```

### **Creating the pencil sketch**

```python
invert_blur_img = 255 - blur_img
pencilsketch = cv2.divide(gray_img, invert_blur_img, scale=256.0)
cv2_imshow(pencilsketch) 
```

### **Displaying the image and pencil sketch**

```python
cv2_imshow(img)
cv2_imshow(pencilsketch) 
```
